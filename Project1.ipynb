```python
import csv
import random
import string

# Function to generate a random name
def generate_name():
    first_name = ''.join(random.choices(string.ascii_uppercase, k=random.randint(4, 10)))
    last_name = ''.join(random.choices(string.ascii_uppercase, k=random.randint(4, 10)))
    return f"{first_name} {last_name}"

# Function to generate a random age between 18 and 80
def generate_age():
    return random.randint(18, 80)

# Function to generate a random city name
def generate_city():
    cities = ['New York', 'San Francisco', 'London', 'Sydney', 'Paris', 'Tokyo']
    return random.choice(cities)

# Function to generate a random email address
def generate_email():
    name = ''.join(random.choices(string.ascii_lowercase, k=random.randint(4, 8)))
    domain = ''.join(random.choices(string.ascii_lowercase, k=random.randint(4, 8)))
    return f"{name}@{domain}.com"

# Function to generate a random phone number
def generate_phone():
    return f"+{random.randint(1, 999)}-{random.randint(100, 999)}-{random.randint(1000, 9999)}"

# Function to generate a random height in centimeters
def generate_height():
    return random.randint(150, 200)

# Function to generate a random weight in kilograms
def generate_weight():
    return random.randint(50, 100)

# Function to generate a random skin texture
def generate_skin_texture():
    skin_textures = ['Smooth', 'Oily', 'Dry', 'Combination', 'Normal']
    return random.choice(skin_textures)

# Function to generate a random skin irritation level
def generate_skin_irritation():
    return random.randint(0, 10)

# Function to generate a random skin sensitivity level
def generate_skin_sensitivity():
    return random.randint(0, 10)

# Function to generate a random skin reaction type
def generate_skin_reaction():
    skin_reactions = ['None', 'Redness', 'Rash', 'Itching', 'Bumps']
    return random.choice(skin_reactions)

# Function to generate a random controlled sebum protection level
def generate_sebum_protection():
    sebum_protection_levels = ['Low', 'Medium', 'High']
    return random.choice(sebum_protection_levels)

# Generate dummy data for control group and experimental group
num_control_group = 500
num_experimental_group = 500

control_data = []
experimental_data = []

for _ in range(num_control_group):
    row = {
        'Group': 'Control',
        'Name': generate_name(),
        'Age': generate_age(),
        'City': generate_city(),
        'Email': generate_email(),
        'Phone': generate_phone(),
        'Height': generate_height(),
        'Weight': generate_weight(),
        'SkinTexture': generate_skin_texture(),
        'SkinIrritation': generate_skin_irritation(),
        'SkinSensitivity': generate_skin_sensitivity(),
        'SkinReaction': generate_skin_reaction(),
        'SebumProtection': generate_sebum_protection(),
    }
    control_data.append(row)

for _ in range(num_experimental_group):
    row = {
        'Group': 'Experimental',
        'Name': generate_name(),
        'Age': generate_age(),
        'City': generate_city(),
        'Email': generate_email(),
        'Phone': generate_phone(),
        'Height': generate_height(),
        'Weight': generate_weight(),
        'SkinTexture': generate_skin_texture(),
        'SkinIrritation': generate_skin_irritation(),
        'SkinSensitivity': generate_skin_sensitivity(),
        'SkinReaction': generate_skin_reaction(),
        'SebumProtection': generate_sebum_protection(),
    }
    experimental_data.append(row)

# Combine both groups into a single dataset
data = control_data + experimental_data

# Save the dataset to a CSV file
csv_file_path = 'comparative_study_dataset.csv'
with open(csv_file_path, mode='w', newline='') as file:
    fieldnames = ['Group', 'Name', 'Age', 'City', 'Email', 'Phone', 'Height', 'Weight',
                  'SkinTexture', 'SkinIrritation', 'SkinSensitivity', 'SkinReaction', 'SebumProtection']
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    # Write header row
    writer.writeheader()

    # Write data rows
    writer.writerows(data)

print(f"{num_control_group} control group rows and {num_experimental_group} experimental group rows created and saved to 'comparative_study_dataset.csv'")
```
